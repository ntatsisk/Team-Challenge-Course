```python
import os
import numpy as np
import copy
import matplotlib.pyplot as plt
np.random.seed(0)
import pandas as pd
import SimpleITK as sitk
import ast
```

```python
# Print the spacing for each patient
spacing = []
image_sizes =[]
patient_ids = []
fpath = './Data'
for subdir in os.listdir(fpath):
    print(subdir)
    patient_id = int(subdir[-3:]) # eg patient085 --> 85
    print(patient_id)
    patient_ids.append(patient_id)
    once = True
    for file in os.listdir(fpath+'/'+subdir):
        if ('frame' in file) and once:
            fullpath = '/'.join((fpath,subdir,file)) 
            im = sitk.ReadImage(fullpath)
            spac = im.GetSpacing()
            image_size = sitk.GetArrayFromImage(im).shape
            spacing.append(spac)
            image_sizes.append(image_size)
            once = False
            print(spac)
            print(image_size)
            

# Create dataframe with info
patient_info = pd.DataFrame([])
patient_info["patient_id"] = patient_ids
patient_info["spacing"] = spacing
image_sizes_np = np.array(image_sizes)
patient_info["image_pixels"] = tuple(image_sizes_np[:,[2,1,0]]) # reverse the order, final: (x,y,z)
```
